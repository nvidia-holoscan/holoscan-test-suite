```python
import cv2
import matplotlib
from matplotlib import pyplot
import numpy as np
import random
```

```python
bytes_per_pixel = 4
pixels_per_line = 1920
lines_per_image = 1080
pixels_per_image = pixels_per_line * lines_per_image
bytes_per_line = pixels_per_line * bytes_per_pixel
bytes_per_image = bytes_per_pixel * pixels_per_image
frames_per_second = 60
```

```python
1920/4*3-250
```

```python
def box(image, sx, sy, w, h, color):
    image[sy:sy+h,sx:sx+w] = color

box_width = 150
box_height = 150
raw_data = np.zeros(bytes_per_image, dtype=np.uint8)
image = raw_data.reshape(lines_per_image, pixels_per_line, bytes_per_pixel)
image[:,:,3] = 255 # set A
for i in range(200):
    sx, sy = random.randrange(pixels_per_line - box_width), random.randrange(lines_per_image - box_height)
    color = [random.randrange(255), random.randrange(255), random.randrange(255), 255]
    box(image, sx, sy, box_width, box_height, color)
u = 450
box(image, 100, 100, u, u, [255, 0, 0, 255])
box(image, 400, 300, u, u, [0, 255, 0, 255])
box(image, 700, 500, u, u, [0, 0, 255, 255])
pyplot.imshow(image)
```

```python
position=(600,200)
font=cv2.FONT_HERSHEY_SIMPLEX
font_size=10
# color is BGR
red=(255, 0, 0, 255)
green=(0, 255, 0, 255)
blue=(0, 0, 255, 255)
thickness=10
lineType=cv2.FILLED
for n in range(5):
    position = (random.randrange(pixels_per_line - 20), random.randrange(lines_per_image - 50))
    cv2.putText(image, "RED", position, font, font_size, color=red, thickness=thickness, lineType=lineType)
    position = (random.randrange(pixels_per_line - 20), random.randrange(lines_per_image - 50))
    cv2.putText(image, "GREEN", position, font, font_size, color=green, thickness=thickness, lineType=lineType)
    position = (random.randrange(pixels_per_line - 20), random.randrange(lines_per_image - 50))
    cv2.putText(image, "BLUE", position, font, font_size, color=blue, thickness=thickness, lineType=lineType)
pyplot.imshow(image)
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGRA2RGBA)
cv2.imwrite("test_image.png", rgb_image)
```
